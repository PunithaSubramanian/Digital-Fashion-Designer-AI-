```python
!pip install diffusers transformers ftfy
from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion pipeline
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe = pipe.to("cuda") # Move the pipeline to the GPU if available
from IPython.display import display
import ipywidgets as widgets

def generate_image(prompt):
  image = pipe(prompt).images[0]  # Generate image from text
  display(image) # Display the generated image

# Create a text input widget
text_input = widgets.Text(
    value="a dress",
    placeholder='Enter text here',
    description='Prompt:',
    disabled=False
)

# Create a button widget
button = widgets.Button(description="Generate Image")

# Link the button to the image generation function
button.on_click(lambda b: generate_image(text_input.value))

# Display the widgets
display(text_input, button)
```
